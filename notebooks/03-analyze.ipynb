```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

# Ordinary List

```python
o_summon_df = pd.read_csv("data/processed/O-2023-02-26.csv", parse_dates=["publication_date"])
```

```python
functions = o_summon_df["function"].unique().tolist()

fig, ax = plt.subplots(figsize=(12, 8))
for function in functions:
    function_df: pd.DataFrame = o_summon_df[(o_summon_df["function"] == function)]
    #function_df: pd.DataFrame = function_df[(function_df["list_full_code"] == "1.30")]
    plot_df = function_df.groupby("publication_date").agg(max_points=pd.NamedAgg(column="points", aggfunc="max")).reset_index()

    # Plot points against publication date
    ax.plot(
        plot_df["publication_date"],
        plot_df["max_points"],
        marker=".",
        linestyle="-",
        label=f"{function}",
    )

ax.set_xlabel("Publication date")
ax.set_ylabel("Points")
ax.set_title("Points over time")
ax.legend()
plt.show()

    
```
